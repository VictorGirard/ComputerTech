```python
import numpy as np
import pandas as dp
import matplotlib as mpl
```

After importing librairies we need to add the data that we wanted to play with

```python
data = dp.read_csv("DataVictorGirard.csv")
nb_row = data.shape[0]
nb_cols = data.shape[1]
print(data)
print(str(nb_row)+" rows")
print(str(nb_cols)+" cols")
```

```python
print("  __  __  __  __  ")
print("Name of cols : ")
for cols in data.columns:
  print(cols)
```

After having opened the data we can try to search for a certain car color, car year... to search with price we need to tweak the price col because now it's containing string (because of the '€')

```python
print("  __  __  __  __  ")
print("Search by color : Orange ")
data.loc[data['color']=="Orange"]
```

```python
print("  __  __  __  __  ")
print("Search by year model : bt 2003 and 2008 ")
(data.loc[data['car_year']>2003]).loc[data['car_year']<2008]
```

Then to play a bit with prices let's modify the content of the price column to be understanded as number

```python
data[data.columns[5:]]= data[data.columns[5:]].apply(lambda prx : prx.str.replace(',','.')).apply(lambda prx: prx.str.replace('€','')).astype(float)
```

```python
print("  __  __  __  __  ")
print("first 10 rows : ")
print(data.iloc[0:10,5]) #select the price column of the 10 first rows
print("Total : ",data.iloc[0:10,5].sum())
print("  __  __  __  __  ")
print("more info : ")
print(data.iloc[0:10,5].describe())
```

```python
print("  __  __  __  __  ")
print("last 10 rows : ")
last_rows = data.iloc[nb_row - 10: nb_row,5]
print(last_rows) 
print("Total : ",last_rows.sum())
print("  __  __  __  __  ")
print("more info : ")
print(last_rows.describe())
```

```python
print("  __  __  __  __  ")
print("sort by color and price : ")
data.sort_values(by=["color","price"])
```
